```python
public class DadJoke
{
    public string Id {get;set;}
    public string Joke {get;set;}
    public int Status {get;set;}
}
```

```python
using System.Text.Json;
JsonSerializerOptions options = new();
options.PropertyNameCaseInsensitive = true;
DadJoke joke = System.Text.Json.JsonSerializer.Deserialize<DadJoke>(json, options);
joke
```

```python
using System.Reflection;

Type reflectedType = joke.GetType();
BindingFlags flags = BindingFlags.Instance | BindingFlags.Public | BindingFlags.GetProperty;

var properties = reflectedType.GetProperties(flags);
properties
```

```python
DadJoke[] jokes = [joke];
DataFrame dfObjects = new();
foreach(var info in properties)
{
    switch(info.PropertyType.Name)
    {
        case "String":
            dfObjects.Columns.Add(new StringDataFrameColumn(info.Name, jokes.Select(j => info.GetValue(j).ToString())));
            break;
        case "Int32":
            dfObjects.Columns.Add(new SingleDataFrameColumn(info.Name, jokes.Select(j => float.Parse(info.GetValue(j).ToString()))));
            break;
    }
}
dfObjects
```
